```python
#เปลี่ยนข้อมูลเป็นตัวเลข
# สมมติว่ามีคอลัมน์ 'Response' ที่ต้องการแปลงจาก 'Yes'/'No' เป็น 1/0
df['อาชีพ'] = df['อาชีพ'].replace({'แพทย์': 0, 'พนักงานบริษัท': 1, 'เจ้าของธุรกิจ': 2, 'รับจ้าง': 3})
df['ประเภทกรมธรรมม์'] = df['ประเภทกรมธรรมม์'].replace({'Single Trip Travel Easy Plus WWE US&Canada':1, 'Single Trip Asia':2 ,'Single_Trip_Travel_Easy_Plus_WWi_US&Canada':3, 'Annual-Trip':4,'Single Trip Asia':5,'Single-Trip-Only':5})
df['แผนประกันภัย']=df['แผนประกันภัย'].replace({'Easy ViSA Plus':1, 'Easy 1':2 ,'Easy 2':3,'Easy 3':4, 'Easy VISA':5})
df['ประเภทของการเคลม']=df['ประเภทของการเคลม'].replace({'ค่ารักษาพยาบาลจากการบาดเจ็บหรือเจ็บป่วย':1, 'การล่าช้าของเที่ยวบิน':2 ,'ความสูญเสียหรือความเสียหายของกระเป๋าเดินทาง':3})
df['ฉ้อฉล/ไม่ฉ้อฉล']=df['ฉ้อฉล/ไม่ฉ้อฉล'].replace({'ฉ้อฉล':1, 'ไม่ฉ้อฉล':2})
df
```

```python
X_train_numeric = X_train.select_dtypes(include=["number"])  # เลือกเฉพาะตัวเลข #เฉพาะ x
```

```python
Q1 =X_train.quantile(0.25)
Q3 = X_train.quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_inner_fence = Q1 - 1.5 * IQR
upper_inner_fence = Q3 + 1.5 * IQR
lower_outer_fence = Q1 - 3 * IQR
upper_outer_fence = Q3 + 3 * IQR
```

```python
import numpy as np #ลบและเติมค่า outlier
# ✅ ลบแถวที่มี Outliers รุนแรง (> 3*IQR)
X_train_cleaned = X_train_numeric[~((X_train_numeric < lower_outer_fence) | (X_train_numeric > upper_outer_fence)).any(axis=1)]

# ✅ แทนค่าที่เป็น Outliers ปกติ (1.5-3 เท่าของ IQR) ด้วยค่ามัธยฐาน
X_train_cleaned = X_train_cleaned.apply(lambda x: np.where((x < lower_inner_fence[x.name]) | (x > upper_inner_fence[x.name]), x.median(), x))
```

```python
outliers_remaining = ((X_train_cleaned < lower_inner_fence) | (X_train_cleaned > upper_inner_fence)).sum()
print("จำนวน Outliers ที่เหลืออยู่ในแต่ละคอลัมน์:\n", outliers_remaining)
```
