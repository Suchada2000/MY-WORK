```python
#ลองเปลี่ยน
import pandas as pd
import numpy as np
from category_encoders import TargetEncoder
from sklearn.model_selection import KFold, train_test_split

# 📌 **ระบุชื่อคอลัมน์ของ Target**
target_column = "Policy_Class"  # เปลี่ยนเป็นชื่อคอลัมน์ที่ใช้จริง

df_plan_not_missing[target_column] = df_plan_not_missing[target_column].astype(str)  # แปลงเป็นข้อความทั้งหมด
df_plan_not_missing['Target_Numeric'] = df_plan_not_missing[target_column].astype('category').cat.codes  # แปลงเป็นตัวเลข

# ลบคลาสที่มีจำนวนตัวอย่าง 1
df_filtered = df_plan_not_missing[df_plan_not_missing[target_column].map(df_plan_not_missing[target_column].value_counts()) > 1]

# แบ่งข้อมูล Train-Test ใหม่
train_df, test_df = train_test_split(df_filtered, test_size=0.2, random_state=42, stratify=df_filtered[target_column])

# 📌 **แปลง Target เป็นตัวเลข (ถ้าเป็นหมวดหมู่)**
train_df['Target_Numeric'] = train_df[target_column].astype('category').cat.codes
test_df['Target_Numeric'] = test_df[target_column].astype('category').cat.codes  # ใช้ไว้เพื่อ Mapping

# เลือกคอลัมน์ที่เป็นหมวดหมู่ (รวมถึง Plan_)
categorical_features = ['Policy_Class'] + train_df.select_dtypes(include=['object']).columns.tolist()
categorical_features.remove(target_column)  # เอา Target ออก

# 📌 **ตั้งค่า K-Folds**
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 📌 **สร้าง DataFrame ใหม่เก็บผลลัพธ์**
train_encoded = train_df.copy()
test_encoded = test_df.copy()

# ทำ Target Encoding สำหรับฟีเจอร์ที่เลือก
for feature in categorical_features:
    train_encoded[f'{feature}_encoded'] = 0

    for train_idx, val_idx in kf.split(train_df):
        fold_train, fold_val = train_df.iloc[train_idx], train_df.iloc[val_idx]

        # Fit TargetEncoder กับ Training Data ของ Fold นั้น ๆ
        encoder = TargetEncoder(cols=[feature])
        encoder.fit(fold_train[feature], fold_train['Target_Numeric'])

        # ใช้ loc แทนค่าของ fold_val ในคอลัมน์ใหม่
        train_encoded.iloc[val_idx, train_encoded.columns.get_loc(f'{feature}_encoded')] = encoder.transform(fold_val[feature])


train_encoded = train_encoded.reset_index(drop=True)

# 📌 **Fit TargetEncoder กับ Train แล้วใช้ Transform Test**
for feature in categorical_features:
    encoder = TargetEncoder(cols=[feature])
    encoder.fit(train_df[feature], train_df['Target_Numeric'])
    test_encoded[f'{feature}_encoded'] = encoder.transform(test_df[feature])

# 📌 **ลบคอลัมน์ Target_Numeric ที่สร้างไว้**
train_encoded = train_encoded.drop(columns=['Target_Numeric'])
test_encoded = test_encoded.drop(columns=['Target_Numeric'])

```
