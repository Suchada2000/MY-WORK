```python
#จัดการกับค่า outlier ด้วยการแทนค่าด้วยค่าเฉลี่ย ควรทำแค่ชุดฝึก
import pandas as pd
import numpy as np

# ฟังก์ชั่นในการหาค่า IQR
def replace_outliers_with_mean(X_train):
    Q1 = X_train.quantile(0.25)
    Q3 = X_train.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # แทนค่า outlier ด้วยค่าเฉลี่ย
    X_train_no_outliers = X_train.apply(lambda x: x.apply(lambda y: x.mean() if y < lower_bound[x.name] or y > upper_bound[x.name] else y))

    return X_train_no_outliers

# ใช้งานฟังก์ชั่น
X_train_cleaned = replace_outliers_with_mean(X_train)
print(X_train_cleaned)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. ทำการ over-sampling ด้วย SMOTE
smote = SMOTE(sampling_strategy='minority', random_state=42)
X_sm, y_sm = smote.fit_resample(X_scaled, y_train) #เก่า # X_scaled, y_train ต้องเท่ากัน

# 2. แบ่งข้อมูลเป็น train และ test (ถ้ายังไม่ได้แบ่ง)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_sm, y_sm, test_size=0.2, random_state=42)


#...............
from sklearn.ensemble import RandomForestClassifier

# สร้างโมเดล Random Forest
modelRandomForestClassifier = RandomForestClassifier(
    n_estimators=100,         # ใช้ 100 ต้นไม้ (เพิ่มหรือลดได้)
    max_depth=5,              # จำกัดความลึกของแต่ละต้นไม้เพื่อลด Overfitting
    min_samples_split=10,     # ต้องมีอย่างน้อย 10 ตัวอย่างก่อนที่จะแยก Node
    min_samples_leaf=5,       # ต้องมีข้อมูลอย่างน้อย 5 ตัวอย่างต่อ Leaf
    class_weight='balanced',  # จัดการปัญหา Class Imbalance โดยให้น้ำหนักแต่ละ Class เท่าๆ กัน
    random_state=42,          # ตั้งค่าให้ผลลัพธ์เหมือนเดิมทุกครั้งที่รัน
    n_jobs=-1                 # ใช้ทุก CPU Core เพื่อเพิ่มความเร็ว
)

# ฝึกโมเดลด้วยชุดข้อมูลที่ผ่าน SMOTE (ถ้าใช้ SMOTE) หรือชุด Train ปกติ
modelRandomForestClassifier.fit(X_sm, y_sm)

# 4. ทำนายผล
y_pred = modelRandomForestClassifier.predict(X_sm)

# 5. ประเมินผลด้วยความแม่นยำ
accuracy = accuracy_score(y_sm, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

from sklearn.metrics import classification_report

print(classification_report(y_sm, y_pred, target_names=["Not Fraud", "Fraud"]))


# 6. แสดง confusion matrix
conf_matrix = confusion_matrix(y_sm, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. ทำการ over-sampling ด้วย SMOTE
smote = SMOTE(sampling_strategy='minority', random_state=42)
X_sm, y_sm = smote.fit_resample(X_scaled, y_train)  # X_scaled, y_train ต้องเท่ากัน

# 2. แบ่งข้อมูลเป็น train และ test
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_sm, y_sm, test_size=0.2, random_state=42)

# 3. สร้างโมเดล Decision Tree
modelDT = DecisionTreeClassifier(
    criterion='gini',        # ใช้ Gini impurity (สามารถเปลี่ยนเป็น 'entropy' ได้)
    max_depth=5,             # จำกัดความลึกของต้นไม้เพื่อป้องกัน overfitting
    random_state=42          # ค่าคงที่เพื่อให้ผลลัพธ์เหมือนกันทุกครั้ง
)

# 4. ฝึกโมเดล Decision Tree
modelDT.fit(X_train1, y_train1)

# 5. ทำนายผล
y_pred = modelDT.predict(X_test1)

# 6. ประเมินผลด้วยความแม่นยำ
accuracy = accuracy_score(y_test1, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# 7. แสดง classification report
print(classification_report(y_test1, y_pred, target_names=["Not Fraud", "Fraud"]))

# 8. แสดง confusion matrix
conf_matrix = confusion_matrix(y_test1, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
