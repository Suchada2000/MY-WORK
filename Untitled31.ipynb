```python
#นำข้อมูลเข้า
import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('/content/sample_data/ข้อมูลตัวอย่าง.xlsx', engine='openpyxl')

# ดูข้อมูลในไฟล์ Excel
print(df.head())  # แสดงข้อมูล 5 แถวแรกเพื่อดูตัวอย่าง
```

```python

#เปลี่ยนข้อมูลเป็นตัวเลข
# สมมติว่ามีคอลัมน์ 'Response' ที่ต้องการแปลงจาก 'Yes'/'No' เป็น 1/0
df['อาชีพ'] = df['อาชีพ'].replace({'แพทย์': 0, 'พนักงานบริษัท': 1, 'เจ้าของธุรกิจ': 2, 'รับจ้าง': 3})
df['ประเภทกรมธรรมม์'] = df['ประเภทกรมธรรมม์'].replace({'Single Trip Travel Easy Plus WWE US&Canada':1, 'Single Trip Asia':2 ,'Single_Trip_Travel_Easy_Plus_WWi_US&Canada':3, 'Annual-Trip':4,'Single Trip Asia':5,'Single-Trip-Only':5})
df['แผนประกันภัย']=df['แผนประกันภัย'].replace({'Easy ViSA Plus':1, 'Easy 1':2 ,'Easy 2':3,'Easy 3':4, 'Easy VISA':5})
df['ประเภทของการเคลม']=df['ประเภทของการเคลม'].replace({'ค่ารักษาพยาบาลจากการบาดเจ็บหรือเจ็บป่วย':1, 'การล่าช้าของเที่ยวบิน':2 ,'ความสูญเสียหรือความเสียหายของกระเป๋าเดินทาง':3})
df['ฉ้อฉล/ไม่ฉ้อฉล']=df['ฉ้อฉล/ไม่ฉ้อฉล'].replace({'ฉ้อฉล':1, 'ไม่ฉ้อฉล':2})
df
```

```python
#เปลี่ยนวันที่ให้เหลือแค่วันสำหรับ nn
from sklearn import preprocessing

# สมมติว่า df เป็น DataFrame ที่มีคอลัมน์ 'date1', 'date2', 'date3' เป็นคอลัมน์วันที่
df['วันที่ซื้อกรมณรรม์'] = pd.to_datetime(df['วันที่ซื้อกรมณรรม์'], format='%d/%m/%Y')
df['วันที่เริ่มคุ้มครอง'] = pd.to_datetime(df['วันที่เริ่มคุ้มครอง'], format='%d/%m/%Y')
df['วันที่สิ้นสุด'] = pd.to_datetime(df['วันที่สิ้นสุด'], format='%d/%m/%Y')
df['วันที่แจ้งเคลม'] = pd.to_datetime(df['วันที่แจ้งเคลม'], format='%d/%m/%Y')
df['วันที่เกิดเหตุ'] = pd.to_datetime(df['วันที่เกิดเหตุ'], format='%d/%m/%Y')

# แปลงวันที่ในแต่ละคอลัมน์เป็นจำนวนวันที่ตั้งแต่ epoch
df['date1_numeric'] = df['วันที่ซื้อกรมณรรม์'].apply(lambda x: x.timestamp())
df['date2_numeric'] = df['วันที่เริ่มคุ้มครอง'].apply(lambda x: x.timestamp())
df['date3_numeric'] = df['วันที่สิ้นสุด'].apply(lambda x: x.timestamp())
df['date4_numeric'] = df['วันที่แจ้งเคลม'].apply(lambda x: x.timestamp())
df['date5_numeric'] = df['วันที่เกิดเหตุ'].apply(lambda x: x.timestamp())

# ลบคอลัมน์วันที่ออกไปจาก X
X = df.drop(['ฉ้อฉล/ไม่ฉ้อฉล','วันที่ซื้อกรมณรรม์', 'วันที่เริ่มคุ้มครอง', 'วันที่สิ้นสุด', 'วันที่แจ้งเคลม','วันที่เกิดเหตุ'], axis=1)
y = df['ฉ้อฉล/ไม่ฉ้อฉล']

df.head()
```

```python
#แยกชุดฝึกและชุดทดสอบ
from sklearn.model_selection import train_test_split

# แบ่งข้อมูลออกเป็น Training set (80%) และ Test set (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# แสดงผลลัพธ์
print("Training data (X_train):")
print(X_train.head())  # ดูข้อมูล 5 แถวแรกของ Training data
print("Test data (X_test):")
print(X_test.head())  # ดูข้อมูล 5 แถวแรกของ Test data
```

```python
#จัดการกับค่า outlier ด้วยการแทนค่าด้วยค่าเฉลี่ย ควรทำแค่ชุดฝึก
import pandas as pd
import numpy as np

# ฟังก์ชั่นในการหาค่า IQR
def replace_outliers_with_mean(X_train):
    Q1 = X_train.quantile(0.25)
    Q3 = X_train.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # แทนค่า outlier ด้วยค่าเฉลี่ย
    X_train_no_outliers = X_train.apply(lambda x: x.apply(lambda y: x.mean() if y < lower_bound[x.name] or y > upper_bound[x.name] else y))

    return X_train_no_outliers

# ใช้งานฟังก์ชั่น
X_train_cleaned = replace_outliers_with_mean(X_train)
print(X_train_cleaned)
```

```python
# คำนวณค่า IQR จัดการกับค่า outlier ด้วยการแทนค่าด้วยค่าเฉลี่ย ควรทำแค่ชุดฝึก
Q1 = y_train.quantile(0.25)
Q3 = y_train.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# แทนค่า outlier ด้วยค่าเฉลี่ย
y_train_no_outliers = y_train.apply(lambda x: y_train.mean() if x < lower_bound or x > upper_bound else x)

# แสดงผลลัพธ์
print(y_train_no_outliers)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. ทำการ over-sampling ด้วย SMOTE
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X_scaled, y_train)

# 2. แบ่งข้อมูลเป็น train และ test (ถ้ายังไม่ได้แบ่ง)
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=42)

# 3. สร้างและฝึกโมเดล
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 4. ทำนายผล
y_pred = model.predict(X_test)

# 5. ประเมินผลด้วยความแม่นยำ
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

# 6. แสดง confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
