```python
#นำข้อมูลเข้า
import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('/content/sample_data/ข้อมูลตัวอย่าง.xlsx', engine='openpyxl')

# ดูข้อมูลในไฟล์ Excel
print(df.head())  # แสดงข้อมูล 5 แถวแรกเพื่อดูตัวอย่าง
```

```python

#เปลี่ยนข้อมูลเป็นตัวเลข
# สมมติว่ามีคอลัมน์ 'Response' ที่ต้องการแปลงจาก 'Yes'/'No' เป็น 1/0
df['อาชีพ'] = df['อาชีพ'].replace({'แพทย์': 0, 'พนักงานบริษัท': 1, 'เจ้าของธุรกิจ': 2, 'รับจ้าง': 3})
df['ประเภทกรมธรรมม์'] = df['ประเภทกรมธรรมม์'].replace({'Single Trip Travel Easy Plus WWE US&Canada':1, 'Single Trip Asia':2 ,'Single_Trip_Travel_Easy_Plus_WWi_US&Canada':3, 'Annual-Trip':4,'Single Trip Asia':5,'Single-Trip-Only':5})
df['แผนประกันภัย']=df['แผนประกันภัย'].replace({'Easy ViSA Plus':1, 'Easy 1':2 ,'Easy 2':3,'Easy 3':4, 'Easy VISA':5})
df['ประเภทของการเคลม']=df['ประเภทของการเคลม'].replace({'ค่ารักษาพยาบาลจากการบาดเจ็บหรือเจ็บป่วย':1, 'การล่าช้าของเที่ยวบิน':2 ,'ความสูญเสียหรือความเสียหายของกระเป๋าเดินทาง':3})
df['ฉ้อฉล/ไม่ฉ้อฉล']=df['ฉ้อฉล/ไม่ฉ้อฉล'].replace({'ฉ้อฉล':1, 'ไม่ฉ้อฉล':2})
df
```

```python
#จัดการกับค่า outlier ด้วยการแทนค่าด้วยค่าเฉลี่ย ควรทำแค่ชุดฝึก
import pandas as pd
import numpy as np

# ฟังก์ชั่นในการหาค่า IQR
def replace_outliers_with_mean(X_train):
    Q1 = X_train.quantile(0.25)
    Q3 = X_train.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # แทนค่า outlier ด้วยค่าเฉลี่ย
    X_train_no_outliers = X_train.apply(lambda x: x.apply(lambda y: x.mean() if y < lower_bound[x.name] or y > upper_bound[x.name] else y))

    return X_train_no_outliers

# ใช้งานฟังก์ชั่น
X_train_cleaned = replace_outliers_with_mean(X_train)
print(X_train_cleaned)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. ทำการ over-sampling ด้วย SMOTE
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X_scaled, y_train)

# 2. แบ่งข้อมูลเป็น train และ test (ถ้ายังไม่ได้แบ่ง)
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=42)

# 3. สร้างและฝึกโมเดล
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 4. ทำนายผล
y_pred = model.predict(X_test)

# 5. ประเมินผลด้วยความแม่นยำ
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

# 6. แสดง confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from flask import Flask, request, jsonify
import joblib
import numpy as np

# โหลดโมเดลที่บันทึกไว้
model = joblib.load('random_forest_model.pkl')

# สร้างแอปพลิเคชัน Flask
app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
    # รับข้อมูลจากผู้ใช้ที่ส่งมาในรูปแบบ JSON
    data = request.get_json(force=True)

    # ดึงข้อมูลที่จำเป็น (เช่น features)
    features = np.array(data['features']).reshape(1, -1)

    # ทำนายผลจากโมเดล
    prediction = model.predict(features)

    # ส่งผลลัพธ์กลับไปเป็น JSON
    return jsonify({'prediction': int(prediction[0])})

if __name__ == '__main__':
    app.run(debug=True)
```
