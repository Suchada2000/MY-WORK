```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# สมมติว่า DataFrame มีคอลัมน์ที่เป็น features และ label (เช่น 'Feature1', 'Feature2', ..., 'Label')
# แยกข้อมูลเป็น X (features) และ y (labels)
# สมมติว่า 'target' คือชื่อของคอลัมน์ที่เป็น target
X = df.drop(columns=['Plan_'])  # ลบคอลัมน์ 'target' ออกจาก DataFrame
y_onehot_array = df['Plan_']  # เก็บคอลัมน์ 'target' ไว้ใน y

# แปลงค่าหมวดหมู่เป็นตัวเลข (ถ้าจำเป็น)
label_encoder = LabelEncoder()
y_onehot_array = label_encoder.fit_transform(y_onehot_array)

# แบ่งข้อมูลเป็น train และ test
X_train, X_test, y_train, y_test = train_test_split(X,y_onehot_array , test_size=0.2, random_state=42)

# สร้าง KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5,
                           algorithm='auto',  # วิธีค้นหาข้อมูลใกล้เคียง
                           metric='euclidean' # วิธีคำนวณระยะห่าง
                          )

# ฝึกโมเดล
knn.fit(X_train, y_train)

# ทำนายผล
y_pred = knn.predict(X_test)

# ประเมินผลลัพธ์
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
# แปลงข้อมูลในคอลัมน์ 'Date' ให้เป็น datetime
df['INCDTE'] = pd.to_datetime(df['INCDTE'], format='%Y%m%d')
df['ExpireDate'] = pd.to_datetime(df['ExpireDate'], format='%Y%m%d')
df['DATREP'] = pd.to_datetime(df['DATREP'], format='%Y%m%d')
df['DATOCC'] = pd.to_datetime(df['DATOCC'], format='%Y%m%d')
```

```python
# แปลงคอลัมน์วันที่ให้เป็น datetime (สมมติว่ามีคอลัมน์ 'INCDTE' และ 'DATOCC') วันที่เริ่มและวันที่เกิดเหตุ
df['DATESALE2'] = pd.to_datetime(df['DATESALE2'])
df['INCDTE'] = pd.to_datetime(df['INCDTE'])
df['DATOCC'] = pd.to_datetime(df['DATOCC'])

# คำนวณระยะห่างระหว่าง 'Start_Date' และ 'End_Date'
df['Date_Difference1'] = abs(df['DATESALE2'] - df['INCDTE']).dt.days # วันเกิดเหตุกับวันที่รายงาน
df['Date_Difference2'] = abs(df['INCDTE'] - df['DATOCC']).dt.days # วันที่เริ่มกับวันที่เกิดเหตุ

# แสดงผลลัพธ์
print(df[['Date_Difference1','Date_Difference2']])
df.head()
```

```python
# กำหนดระดับความเสี่ยงตาม Claim_Count
df['Claim_Risk_Level'] = df['Claim_Count'].apply(lambda x:
    'ไม่มีความเสี่ยง' if x < 4 else
    'มีความเสี่ยงน้อย' if x == 4 else
    'มีความเสี่ยงปานกลาง' if x == 5 else
    'มีความเสี่ยงมาก'
)

# แสดงผลลัพธ์บางส่วน
print(df[['Dummy ID', 'Claim_Count', 'Claim_Risk_Level']].head())
```

```python
# นับจำนวน Claim ตาม Dummy ID และ ReserveDescription
claim_counts_per_customer = df.groupby(['Dummy ID', 'ReserveDescription']).size().reset_index(name='Claim_Count')

# รวมค่า Claim_Count กลับเข้า df
df = df.merge(claim_counts_per_customer, on=['Dummy ID', 'ReserveDescription'], how='left')

# กำหนดระดับความเสี่ยงตาม Claim_Count
df['Claim_Risk_Level_A'] = df['Claim_Count'].apply(lambda x:
    'ไม่มีความเสี่ยง' if x < 3 else
    'มีความเสี่ยงน้อย' if x == 3 else
    'มีความเสี่ยงปานกลาง' if x == 4 else
    'มีความเสี่ยงมาก'
)

# แสดงผลลัพธ์บางส่วน
print(df[['Dummy ID', 'ReserveDescription', 'Claim_Count', 'Claim_Risk_Level_A']].head())
```
