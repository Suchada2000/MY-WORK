```python
# แปลงคอลัมน์ 'DateTime' ให้เป็น datetime แล้วตัดเวลาออก
df['DATESALE'] = pd.to_datetime(df['DATESALE'], errors='coerce')
df['DATESALE2'] = df['DATESALE'].dt.date
# แสดงผลลัพธ์
# ลบคอลัมน์ 'Date_Difference'
df = df.drop(columns=['DATESALE'])
df.head()
```

```python
# แปลงข้อมูลในคอลัมน์ 'Date' ให้เป็น datetime
df['INCDTE'] = pd.to_datetime(df['INCDTE'], format='%Y%m%d')
df['ExpireDate'] = pd.to_datetime(df['ExpireDate'], format='%Y%m%d')
df['DATREP'] = pd.to_datetime(df['DATREP'], format='%Y%m%d')
df['DATOCC'] = pd.to_datetime(df['DATOCC'], format='%Y%m%d')

# แสดงผลลัพธ์
df.head()

```

```python
# แปลงคอลัมน์วันที่ให้เป็น datetime (สมมติว่ามีคอลัมน์ 'INCDTE' และ 'DATOCC') วันที่เริ่มและวันที่เกิดเหตุ
df['INCDTE'] = pd.to_datetime(df['INCDTE'])
df['DATREP'] = pd.to_datetime(df['DATREP'])
df['DATOCC'] = pd.to_datetime(df['DATOCC'])
df['ExpireDate'] = pd.to_datetime(df['ExpireDate'])

# คำนวณระยะห่างระหว่าง 'Start_Date' และ 'End_Date'
df['Date_Difference1'] = abs(df['DATOCC'] - df['DATREP']).dt.days # วันเกิดเหตุกับวันที่รายงาน
df['Date_Difference2'] = abs(df['INCDTE'] - df['DATOCC']).dt.days # วันที่เริ่มกับวันที่เกิดเหตุ
df['Date_Difference3'] = abs(df['DATOCC'] - df['ExpireDate']).dt.days # วันที่เกิดกับวันที่หมดอายุ

# แสดงผลลัพธ์
print(df[['Date_Difference1','Date_Difference2','Date_Difference3']])
df.head()
```

```python
df.to_excel('date_differences.xlsx', index=False)
```

```python
average_date_diff_1 = df['Date_Difference1'].mean()
average_date_diff_2 = df['Date_Difference2'].mean()
average_date_diff_3 = df['Date_Difference3'].mean()

# แสดงค่าเฉลี่ย
print(f"ค่าเฉลี่ยของ Date_Difference1: {average_date_diff_1}")
print(f"ค่าเฉลี่ยของ Date_Difference2: {average_date_diff_2}")
print(f"ค่าเฉลี่ยของ Date_Difference3: {average_date_diff_3}")
```

```python
# คำนวณร้อยละจากค่าเฉลี่ย (อ้างอิงจากผลรวมของค่าทั้งหมดในคอลัมน์ Date_Difference1)
total1 = df['Date_Difference1'].sum()
total2 = df['Date_Difference2'].sum()
total3 = df['Date_Difference3'].sum()

percentage_diff_1 = (average_date_diff_1 / total1) * 100
percentage_diff_2 = (average_date_diff_2 / total2) * 100
percentage_diff_3 = (average_date_diff_3 / total3) * 100

# แสดงผลลัพธ์
print(f"ค่าเฉลี่ยของ Date_Difference1: {average_date_diff_1} - ร้อยละ: {percentage_diff_1}%")
print(f"ค่าเฉลี่ยของ Date_Difference2: {average_date_diff_2} - ร้อยละ: {percentage_diff_2}%")
print(f"ค่าเฉลี่ยของ Date_Difference3: {average_date_diff_3} - ร้อยละ: {percentage_diff_3}%")
```
