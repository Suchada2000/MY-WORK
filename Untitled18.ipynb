```python
#train and test set split, and feature scaling
#แยกชุดฝึกและทดสอบ 40%และ60% random_state=0 กำหนด seed สำหรับการสุ่ม เพื่อให้การสุ่มเดิมได้ผลเหมือนกันทุกครั้ง (สำหรับการทำซ้ำ)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = 0)
sc = StandardScaler()
#นำเข้าคลาส StandardScaler เพื่อปรับข้อมูลให้อยู่ในรูปแบบที่มีค่าเฉลี่ย (mean) เป็น 0 และส่วนเบี่ยงเบนมาตรฐาน (standard deviation) เป็น 1
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Keras เป็นไลบรารีสำหรับการพัฒนาโมเดล Deep Learning ซึ่งได้รับความนิยมสูงเนื่องจากมีความใช้งานง่ายและมีความยืดหยุ่นสูง
import keras
from keras.models import Sequential
from keras.layers import Dense #to add layers

#there is no rule on how many nodes each hidden layer should have
classifier = Sequential()
classifier.add(Dense(units = 11, kernel_initializer = 'uniform', activation = 'relu', input_dim = 20))
#กำหนดการสุ่มค่าเริ่มต้นของน้ำหนัก (weights) จากการแจกแจงแบบ Uniform Distribution
#ฟังก์ชันการกระตุ้น (activation function) ใช้ ReLU (Rectified Linear Unit) ซึ่งเหมาะสำหรับเลเยอร์ซ่อน

#เลเยอร์ถัดมา
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
#input_dim --> remove it as it already knows what to expect.

#the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
# ฟังก์ชัน sigmoid เหมาะสำหรับปัญหาไบนารี เนื่องจากจะคืนค่าความน่าจะเป็นระหว่าง 0 และ 1 เช่น คอลัมเพศ 1 โหนด มี 2 คลาส

#Adam Optimizer ซึ่งเป็นเทคนิคที่ผสมผสาน Stochastic Gradient Descent (SGD) และการปรับอัตราการเรียนรู้แบบอัตโนมัติ
classifier.compile(optimizer = 'adam', loss = "binary_crossentropy", metrics = ['accuracy'])
#optimizer = algorithm to find the optimal set of weights in ANN
#loss = functions that should be optimized. if more than two categories, use "categorical_crossentropy"
#metrics = criterion used to calculate the performance of the model.
```

```python
classifier.fit(X_train, Y_train, batch_size = 32, epochs = 20)
#หาก training loss ลดลงแต่ validation loss เพิ่มขึ้นหลังจากจำนวน epoch ที่มากขึ้น แสดงว่าโมเดลเริ่ม overfitting
```

```python
#predicting the results

Y_pred = classifier.predict(X_test)
Y_pred = (Y_pred > 0.5) #to classify each probability into True or False

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print (cm, '\n\n', Y_pred[:5, :])
#True Positives (TP): 30 — จำนวนที่ทำนายว่าเป็น True และจริงคือ True .............. (FP): 10 — จำนวนที่ทำนายว่าเป็น True แต่จริงคือ False.................False Negatives (FN): 0 — จำนวนที่ทำนายว่าเป็น False แต่จริงคือ True.....................True Negatives (TN): 0 — จำนวนที่ทำนายว่าเป็น False และจริงคือ False
```
