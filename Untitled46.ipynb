```python

import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report

# อ่านไฟล์ Excel
df = pd.read_excel('/content/sample_data/originalview10.xlsx')
# กำหนดว่าถ้าค่า Risk_Level เป็น "High" หรือ "Very High" ให้ถือว่าเป็น Fraud
df['Predicted_Fraud'] = df['Overall_Risk_Level'].apply(lambda x: 1 if x in ['ความเสี่ยงสูงมาก'] else 0)
df['Suspicious_Flag'] = df['Suspicious_Risk'].apply(lambda x: 1 if x == 'Suspicious' else 0)
df_selected = df[['ID_Card','Predicted_Fraud', 'Overall_Risk_Level','Suspicious_Risk','Suspicious_Flag']]
# คำนวณ Confusion Matrix
tn, fp, fn, tp = confusion_matrix(df['Suspicious_Flag'], df['Predicted_Fraud']).ravel()

print(f"True Positive (จับ Fraud ได้ถูกต้อง): {tp}")
print(f"False Positive (ระบุผิดว่าเป็น Fraud): {fp}")
print(f"False Negative (พลาด Fraud จริง): {fn}")
print(f"True Negative (ระบุ Non-Fraud ถูกต้อง): {tn}")

# แสดงรายงานประสิทธิภาพโมเดล
print("\nรายงานประสิทธิภาพโมเดล:")
print(classification_report(df['Suspicious_Flag'], df['Predicted_Fraud']))
```
