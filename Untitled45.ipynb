```python

import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('/content/sample_data/Untitled spreadsheet.xlsx')
```

```python

# แปลงคอลัมน์ DATESALE เป็น datetime โดยไม่ระบุ format และจัดการกับข้อผิดพลาด
df['DATESALE'] = pd.to_datetime(df['DATESALE'], errors='coerce')

# แปลงให้เป็นรูปแบบ YYYY-MM-DD
df['DATESALE'] = df['DATESALE'].dt.strftime('%Y-%m-%d')
```

```python
print(df.dtypes)
```

```python

# แปลงคอลัมน์ DATESALE เป็น string แล้วค่อยแปลงเป็น datetime โดยระบุ format
df['INCDTE'] = pd.to_datetime(df['INCDTE'].astype(str), format='%Y%m%d', errors='coerce')

# แปลงให้เป็นรูปแบบ YYYY-MM-DD
df['INCDTE'] = df['INCDTE'].dt.strftime('%Y-%m-%d')
```

```python

# แปลงเป็น string แล้วค่อยแปลงเป็น datetime
df['DATREP'] = pd.to_datetime(df['DATREP'].astype(str), format='%Y%m%d')
# แปลงเป็นรูปแบบ YYYY-MM-DD
df['DATREP'] = df['DATREP'].dt.strftime('%Y-%m-%d')

# แปลงเป็น string แล้วค่อยแปลงเป็น datetime
df['DATOCC'] = pd.to_datetime(df['DATOCC'].astype(str), format='%Y%m%d')
# แปลงเป็นรูปแบบ YYYY-MM-DD
df['DATOCC'] = df['DATOCC'].dt.strftime('%Y-%m-%d')
```

```python

# แปลงเป็น datetime โดยให้ pandas จัดการรูปแบบอัตโนมัติ
df['DATESALE'] = pd.to_datetime(df['DATESALE'], format='mixed', errors='coerce')
df['INCDTE'] = pd.to_datetime(df['INCDTE'], format='mixed', errors='coerce')
```

```python

# แปลงเป็น datetime โดยให้ pandas จัดการรูปแบบอัตโนมัติ
df['DATOCC'] = pd.to_datetime(df['DATOCC'], format='mixed', errors='coerce')
df['DATOCC'] = pd.to_datetime(df['DATOCC'], format='mixed', errors='coerce')
```
