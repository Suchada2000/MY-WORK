```python
import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('/content/sample_data/Untitled Folder/ข้อมูลตัวอย่าง.xlsx', engine='openpyxl')

# ดูข้อมูลในไฟล์ Excel
print(df.head())  # แสดงข้อมูล 5 แถวแรกเพื่อดูตัวอย่าง

# สมมติว่ามีคอลัมน์ 'Response' ที่ต้องการแปลงจาก 'Yes'/'No' เป็น 1/0
df['อาชีพ'] = df['อาชีพ'].replace({'แพทย์': 0, 'พนักงานบริษัท': 1, 'เจ้าของธุรกิจ': 2, 'รับจ้าง': 3})
df['ประเภทกรมธรรมม์'] = df['ประเภทกรมธรรมม์'].replace({'Single Trip Travel Easy Plus WWE US&Canada':1, 'Single Trip Asia':2 ,'Single_Trip_Travel_Easy_Plus_WWi_US&Canada':3, 'Annual-Trip':4,'Single Trip Asia':5})
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. อ่านข้อมูลจากไฟล์ Excel
# ใช้ pandas ในการอ่านไฟล์ Excel
input_file = '/content/sample_data/Untitled Folder/ข้อมูลตัวอย่าง.xlsx'  # ชื่อไฟล์ Excel ของคุณ
df = pd.read_excel(input_file)

# 2. เลือกเฉพาะคอลัมน์ที่ต้องการทำมาตรฐาน (เลือกเฉพาะฟีเจอร์ที่ไม่ใช่ประเภทข้อมูล)
# สมมติว่า df มีคอลัมน์ที่ไม่ใช่ข้อความ (numeric columns) อยู่
numeric_columns = df.select_dtypes(include=['number']).columns

# 3. สร้าง StandardScaler และทำการแปลงข้อมูล
scaler = StandardScaler()

# ใช้ scaler ในการแปลงข้อมูล
df_scaled = df.copy()  # สร้างสำเนาของข้อมูล
df_scaled[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# 4. ตรวจสอบว่า ข้อมูลถูกแปลงเรียบร้อยหรือไม่
print(df_scaled.head())  # แสดงข้อมูลแปลงแล้ว

# 5. ส่งออกข้อมูลที่แปลงแล้วไปยังไฟล์ Excel ใหม่
output_file = 'scaled_data.xlsx'  # ชื่อไฟล์ Excel ที่จะบันทึกข้อมูลแปลงแล้ว
df_scaled.to_excel(output_file, index=False)

print(f"ข้อมูลที่แปลงแล้วถูกบันทึกในไฟล์ {output_file}")
```
