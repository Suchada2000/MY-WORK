```python
#นำข้อมูลเข้าและเปลี่ยนให้เป็นตัวเลข
import pandas as pd

# อ่านไฟล์ Excel
df = pd.read_excel('/content/sample_data/ข้อมูลตัวอย่าง.xlsx', engine='openpyxl')

# ดูข้อมูลในไฟล์ Excel
print(df.head())  # แสดงข้อมูล 5 แถวแรกเพื่อดูตัวอย่าง

# สมมติว่ามีคอลัมน์ 'Response' ที่ต้องการแปลงจาก 'Yes'/'No' เป็น 1/0
df['อาชีพ'] = df['อาชีพ'].replace({'แพทย์': 0, 'พนักงานบริษัท': 1, 'เจ้าของธุรกิจ': 2, 'รับจ้าง': 3})
df['ประเภทกรมธรรมม์'] = df['ประเภทกรมธรรมม์'].replace({'Single Trip Travel Easy Plus WWE US&Canada':1, 'Single Trip Asia':2 ,'Single_Trip_Travel_Easy_Plus_WWi_US&Canada':3, 'Annual-Trip':4,'Single Trip Asia':5,'Single-Trip-Only':5})
df['แผนประกันภัย']=df['แผนประกันภัย'].replace({'Easy ViSA Plus':1, 'Easy 1':2 ,'Easy 2':3,'Easy 3':4, 'Easy VISA':5})
df['ประเภทของการเคลม']=df['ประเภทของการเคลม'].replace({'ค่ารักษาพยาบาลจากการบาดเจ็บหรือเจ็บป่วย':1, 'การล่าช้าของเที่ยวบิน':2 ,'ความสูญเสียหรือความเสียหายของกระเป๋าเดินทาง':3})
df['ฉ้อฉล/ไม่ฉ้อฉล']=df['ฉ้อฉล/ไม่ฉ้อฉล'].replace({'ฉ้อฉล':1, 'ไม่ฉ้อฉล':2})
df
```

# clean+นำเข้า+แบ่ง

```python
#เติมค่าในช่องว่าง
from sklearn.impute import SimpleImputer

# สร้าง SimpleImputer
imputer = SimpleImputer(strategy='mean')  # ใช้การเติมค่าเฉลี่ย

df['ค่าเบี้ยประกันภัย'] = imputer.fit_transform(df[['ค่าเบี้ยประกันภัย']])

print(df)
```

```python
#เปลี่ยนวันที่ให้เหลือแค่วันสำหรับ nn
from sklearn import preprocessing

# สมมติว่า df เป็น DataFrame ที่มีคอลัมน์ 'date1', 'date2', 'date3' เป็นคอลัมน์วันที่
df['วันที่ซื้อกรมณรรม์'] = pd.to_datetime(df['วันที่ซื้อกรมณรรม์'], format='%d/%m/%Y')
df['วันที่เริ่มคุ้มครอง'] = pd.to_datetime(df['วันที่เริ่มคุ้มครอง'], format='%d/%m/%Y')
df['วันที่สิ้นสุด'] = pd.to_datetime(df['วันที่สิ้นสุด'], format='%d/%m/%Y')
df['วันที่แจ้งเคลม'] = pd.to_datetime(df['วันที่แจ้งเคลม'], format='%d/%m/%Y')
df['วันที่เกิดเหตุ'] = pd.to_datetime(df['วันที่เกิดเหตุ'], format='%d/%m/%Y')

# แปลงวันที่ในแต่ละคอลัมน์เป็นจำนวนวันที่ตั้งแต่ epoch
df['date1_numeric'] = df['วันที่ซื้อกรมณรรม์'].apply(lambda x: x.timestamp())
df['date2_numeric'] = df['วันที่เริ่มคุ้มครอง'].apply(lambda x: x.timestamp())
df['date3_numeric'] = df['วันที่สิ้นสุด'].apply(lambda x: x.timestamp())
df['date2_numeric'] = df['วันที่แจ้งเคลม'].apply(lambda x: x.timestamp())
df['date3_numeric'] = df['วันที่เกิดเหตุ'].apply(lambda x: x.timestamp())

# ลบคอลัมน์วันที่ออกไปจาก X
X = df.drop(['ฉ้อฉล/ไม่ฉ้อฉล','วันที่ซื้อกรมณรรม์', 'วันที่เริ่มคุ้มครอง', 'วันที่สิ้นสุด', 'วันที่แจ้งเคลม','วันที่เกิดเหตุ'], axis=1)
y = df['ฉ้อฉล/ไม่ฉ้อฉล']
```

```python
#แยกชุดฝึกและชุดทดสอบ
from sklearn.model_selection import train_test_split

# แบ่งข้อมูลออกเป็น Training set (80%) และ Test set (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# แสดงผลลัพธ์
print("Training data (X_train):")
print(X_train.head())  # ดูข้อมูล 5 แถวแรกของ Training data
print("Test data (X_test):")
print(X_test.head())  # ดูข้อมูล 5 แถวแรกของ Test data
```

```python
# สมมติว่าคุณต้องการตรวจหาค่าผิดปกติในคอลัมน์ 'Column_Name'
column_name = 'จำนวนเงินความเสียหายที่บริษัทจ่ายไป (GRS Paid)'

# คำนวณ Z-Score
mean = df[column_name].mean()
std_dev = df[column_name].std()

# คำนวณ Z-Score สำหรับแต่ละแถว
df['Z-Score'] = (df[column_name] - mean) / std_dev

# ตรวจสอบค่าที่มี Z-Score > 3 หรือ < -3 เป็น outliers
outliers_zscore = df[df['Z-Score'].abs() > 3]

# แสดงข้อมูลที่มี Z-Score > 3 หรือ < -3
print("Outliers based on Z-Score:")
print(outliers_zscore)

# คำนวณ IQR (Interquartile Range)
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1

# คำนวณค่า Lower Bound และ Upper Bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# ตรวจสอบค่าผิดปกติที่อยู่นอกขอบเขต
outliers_iqr = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

# แสดงข้อมูลที่มี outliers จาก IQR
print("Outliers based on IQR:")
print(outliers_iqr)
```

# แบ่งกลุ่ม

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.utils import resample
import numpy as np

# สมมติว่า X_train และ y_train เป็นข้อมูลที่คุณมี
# รวม X_train และ y_train เข้าด้วยกันเพื่อจัดการข้อมูล
data = pd.concat([X_train, y_train], axis=1)

# สมมติว่า y_train มีค่าของ Class 0 และ Class 1
class_0 = X_train[y_train == 0]  # ข้อมูลที่มี class 0
class_1 = X_train[y_train == 1]  # ข้อมูลที่มี class 1

# ตรวจสอบจำนวนข้อมูลที่ได้
print(f'จำนวนข้อมูลใน Class 0: {class_0.shape[0]}')
print(f'จำนวนข้อมูลใน Class 1: {class_1.shape[0]}')

# แยกกลุ่มที่มีจำนวนมาก (Class 0) และกลุ่มที่มีจำนวนน้อย (Class 1)
class_0 = data[data['ฉ้อฉล/ไม่ฉ้อฉล'] == 2]
class_1 = data[data['ฉ้อฉล/ไม่ฉ้อฉล'] == 1]

# 2. ทำการ Clustering ข้อมูลที่เหลือจาก Class 0
X_class_0 = class_0.drop('ฉ้อฉล/ไม่ฉ้อฉล', axis=1)

# ใช้ KMeans เพื่อทำ clustering (ใช้ 2 คลัสเตอร์ในการแยกกลุ่มใหม่)
kmeans = KMeans(n_clusters=2, random_state=42)
class_0['ฉ้อฉล/ไม่ฉ้อฉล'] = kmeans.fit_predict(X_class_0)

# 3. รวมข้อมูลที่ถูกตัดออก (Class 1) กับข้อมูลใหม่ที่ได้จาก Clustering
# เพิ่ม Class 1 กลับไปในข้อมูล
data_class_1 = class_1.copy()

# เปลี่ยนชื่อ 'Target' เป็น 2 สำหรับคลาสที่เพิ่มเข้ามา
data_class_1['ฉ้อฉล/ไม่ฉ้อฉล'] = 2

# รวมข้อมูลทั้งหมด (ทั้งจากการ clustering และข้อมูลที่เพิ่ม)
data_final = pd.concat([class_0, data_class_1])

# แยก X_train และ y_train ออกใหม่
X_train_final = data_final.drop('ฉ้อฉล/ไม่ฉ้อฉล', axis=1)
y_train_final = data_final['ฉ้อฉล/ไม่ฉ้อฉล']

# ผลลัพธ์ที่ได้คือข้อมูลที่มีทั้งหมด 3
#คลาส 1 ไม่ฉ้อ 0 อาจฉ้อ 2 ฉ้อ
```

# สมดุล

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, balanced_accuracy_score

# ใช้ RandomOverSampler เพื่อเพิ่มข้อมูลในคลาสที่มีน้อย
ros = RandomOverSampler(random_state=0)
X_train_balanced, y_train_balanced = ros.fit_resample(X_train_final, y_train_final)

# สร้างโมเดล MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(50, 25), max_iter=5000, activation='relu', random_state=42)

# ฝึกโมเดล
mlp.fit(X_train_balanced, y_train_balanced)

# ทำนายผล
y_pred = mlp.predict(X_test)

# คำนวณค่าความแม่นยำ (Accuracy) และ Balanced Accuracy
accuracy = accuracy_score(y_test, y_pred)
balanced_acc = balanced_accuracy_score(y_test, y_pred)

# แสดงค่าความแม่นยำและรายงาน
print(f"Accuracy Score: {accuracy}")
print(f"Balanced Accuracy Score: {balanced_acc}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.utils import resample
import numpy as np

# สมมติว่า X_train และ y_train เป็นข้อมูลที่คุณมี
# รวม X_train และ y_train เข้าด้วยกันเพื่อจัดการข้อมูล
data = pd.concat([X_train, y_train], axis=1)

# สมมติว่า y_train มีค่าของ Class 0 และ Class 1
class_0 = X_train[y_train == 0]  # ข้อมูลที่มี class 0
class_1 = X_train[y_train == 1]  # ข้อมูลที่มี class 1

# ตรวจสอบจำนวนข้อมูลที่ได้
print(f'จำนวนข้อมูลใน Class 0: {class_0.shape[0]}')
print(f'จำนวนข้อมูลใน Class 1: {class_1.shape[0]}')

# แยกกลุ่มที่มีจำนวนมาก (Class 0) และกลุ่มที่มีจำนวนน้อย (Class 1)
class_0 = data[data['ฉ้อฉล/ไม่ฉ้อฉล'] == 2]
class_1 = data[data['ฉ้อฉล/ไม่ฉ้อฉล'] == 1]

# 2. ทำการ Clustering ข้อมูลที่เหลือจาก Class 0
X_class_0 = class_0.drop('ฉ้อฉล/ไม่ฉ้อฉล', axis=1)

# ใช้ KMeans เพื่อทำ clustering (ใช้ 2 คลัสเตอร์ในการแยกกลุ่มใหม่)
kmeans = KMeans(n_clusters=2, random_state=42)
class_0['ฉ้อฉล/ไม่ฉ้อฉล'] = kmeans.fit_predict(X_class_0)

# 3. รวมข้อมูลที่ถูกตัดออก (Class 1) กับข้อมูลใหม่ที่ได้จาก Clustering
# เพิ่ม Class 1 กลับไปในข้อมูล
data_class_1 = class_1.copy()

# เปลี่ยนชื่อ 'Target' เป็น 2 สำหรับคลาสที่เพิ่มเข้ามา
data_class_1['ฉ้อฉล/ไม่ฉ้อฉล'] = 0

# รวมข้อมูลทั้งหมด (ทั้งจากการ clustering และข้อมูลที่เพิ่ม)
data_final = pd.concat([class_0, data_class_1])

# แยก X_train และ y_train ออกใหม่
X_train_final = data_final.drop('ฉ้อฉล/ไม่ฉ้อฉล', axis=1)
y_train_final = data_final['ฉ้อฉล/ไม่ฉ้อฉล']

# ผลลัพธ์ที่ได้คือข้อมูลที่มีทั้งหมด 3
#คลาส 1 ไม่ฉ้อ 0 อาจฉ้อ 2 ฉ้อ
```

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, balanced_accuracy_score

# ใช้ RandomOverSampler เพื่อเพิ่มข้อมูลในคลาสที่มีน้อย
ros = RandomOverSampler(random_state=0)
X_train_balanced, y_train_balanced = ros.fit_resample(X_train_final, y_train_final)

# สร้างโมเดล MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(50, 25), max_iter=5000, activation='relu', random_state=42)

# ฝึกโมเดล
mlp.fit(X_train_balanced, y_train_balanced)

# ทำนายผล
y_pred = mlp.predict(X_test)

# คำนวณค่าความแม่นยำ (Accuracy) และ Balanced Accuracy
accuracy = accuracy_score(y_test, y_pred)
balanced_acc = balanced_accuracy_score(y_test, y_pred)

# แสดงค่าความแม่นยำและรายงาน
print(f"Accuracy Score: {accuracy}")
print(f"Balanced Accuracy Score: {balanced_acc}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
